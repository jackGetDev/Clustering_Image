```python
import matplotlib.pyplot as plt
from skimage.color import rgb2gray, rgba2rgb
import skimage.io
import numpy as np

# Load the image
data = skimage.io.imread("/content/MRI_blackandwhite.png")

# Convert RGBA to RGB
data_rgb = rgba2rgb(data)

# Convert the image to grayscale
grayscale = rgb2gray(data_rgb)

# Split original and grayscale
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(data_rgb)
ax[0].set_title("Original")
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import skimage.io
import numpy as np
from sklearn.cluster import KMeans

# Load the image
data = skimage.io.imread("/content/MRI_blackandwhite.png")


# Convert RGBA to RGB
data_rgb = rgba2rgb(data)
# Convert the image to grayscale
grayscale = rgb2gray(data_rgb)

# Reshape grayscale image to prepare for KMeans clustering
image_gray = grayscale.reshape(-1, 1)
n_clusters = 6

# Perform KMeans clustering with multiple initializations
kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0)
kmeans.fit(image_gray)
clustered_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Create subplots to display clustered images for each cluster center
fig, axes = plt.subplots(2, 5, figsize=(20, 10))
axes = axes.ravel()

for i in range(n_clusters):
    cluster_i_indices = np.where(labels == i)[0]  # Get the indices of cluster i
    clustered_i = np.copy(image_gray)
    clustered_i[np.setdiff1d(np.arange(len(labels)), cluster_i_indices)] = 1
    clustered_i = clustered_i.reshape(grayscale.shape)

    axes[i].imshow(clustered_i, cmap=plt.cm.gray)
    axes[i].set_title(f"Cluster {i + 1}")

# Hide remaining empty subplots
for i in range(n_clusters, 10):
    axes[i].axis("off")

plt.show()
```
